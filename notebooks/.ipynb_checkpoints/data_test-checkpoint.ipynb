```python
#categorizing the percentage of a race in the STEM field in comparison to that races percentage in the work force

def categorize_white(percent):
    if percent < 67.2: return 'less than pop'
    elif percent == 67.2: return 'equal to pop'
    elif percent > 67.2: return 'more than pop'
    else: return np.nan

total['White Representation'] = total['White Percent'].apply(categorize_white)

def categorize_black(percent):
    if percent < 11: return 'less than pop'
    elif percent == 11: return 'equal to pop'
    elif percent > 11: return 'more than pop'
    else: return np.nan

total['Black Representation'] = total['Black Percent'].apply(categorize_black)

def categorize_asian(percent):
    if percent < 5.9: return 'less than pop'
    elif percent == 5.9: return 'equal to pop'
    elif percent > 5.9: return 'more than pop'
    else: return np.nan

total['Asian Representation'] = total['Asian Percent'].apply(categorize_asian)

def categorize_hispanic(percent):
    if percent < 15.8: return 'less than pop'
    elif percent == 15.8: return 'equal to pop'
    elif percent > 15.8: return 'more than pop'
    else: return np.nan

total['Hispanic Representation'] = total['Hispanic Percent'].apply(categorize_hispanic)
```

```python
#grouping the categorizations

wrep_data = total.groupby('White Representation').size().reset_index().rename(columns={0:'n_wrep'})
display(wrep_data)

brep_data = total.groupby('Black Representation').size().reset_index().rename(columns={0:'n_brep'})
display(brep_data)

arep_data = total.groupby('Asian Representation').size().reset_index().rename(columns={0:'n_arep'})
display(arep_data)

hrep_data = total.groupby('Hispanic Representation').size().reset_index().rename(columns={0:'n_hrep'})
display(hrep_data)
```

```python
# compares percentage of white people in work force with percent in a specific occupation
wrep_data.plot.bar(
    x='White Representation', y='n_wrep',
    legend=None,                 
    rot=45,                                     
    color='#d9e3f0'                             
);

# compares percentage of black people in work force with percent in a specific occupation
brep_data.plot.bar(
    x='Black Representation', y='n_brep',
    legend=None,                                
    rot=45,                                     
    color='#f47373'                             
);

# compares percentage of asian people in work force with percent in a specific occupation
arep_data.plot.bar(
    x='Asian Representation', y='n_arep',
    legend=None,                                
    rot=45,                                     
    color='#37d67a'                            
);

# compares percentage of hispanic people in work force with percent in a specific occupation
hrep_data.plot.bar(
    x='Hispanic Representation', y='n_hrep',
    legend=None,                                
    rot=45,                                     
    color='#ba68c8'                             
);
```

```python
#categorizing the percentage of a race in the STEM field in comparison to that races percentage in earning STEM degrees

def categorize_white_degree(percent):
    if percent < 62.7: return 'less than degree'
    elif percent == 62.7: return 'equal to degree'
    elif percent > 62.7: return 'more than degree'
    else: return np.nan

total['White Degree'] = total['White Percent'].apply(categorize_white_degree)

def categorize_black_degree(percent):
    if percent < 8.8: return 'less than degree'
    elif percent == 8.8: return 'equal to degree'
    elif percent > 8.8: return 'more than degree'
    else: return np.nan

total['Black Degree'] = total['Black Percent'].apply(categorize_black_degree)

def categorize_asian_degree(percent):
    if percent < 9.7: return 'less than degree'
    elif percent == 9.7: return 'equal to degree'
    elif percent > 9.7: return 'more than degree'
    else: return np.nan

total['Asian Degree'] = total['Asian Percent'].apply(categorize_asian_degree)

def categorize_hispanic_degree(percent):
    if percent < 10.3: return 'less than degree'
    elif percent == 10.3: return 'equal to degree'
    elif percent > 10.3: return 'more than degree'
    else: return np.nan

total['Hispanic Degree'] = total['Hispanic Percent'].apply(categorize_hispanic_degree)
```

```python
#grouping the categorizations

wdeg_data = total.groupby('White Degree').size().reset_index().rename(columns={0:'n_wdeg'})
display(wdeg_data)

bdeg_data = total.groupby('Black Degree').size().reset_index().rename(columns={0:'n_bdeg'})
display(bdeg_data)

adeg_data = total.groupby('Asian Degree').size().reset_index().rename(columns={0:'n_adeg'})
display(adeg_data)

hdeg_data = total.groupby('Hispanic Degree').size().reset_index().rename(columns={0:'n_hdeg'})
display(hdeg_data)
```

```python
# compares percentage of white people in work force with percent in a specific occupation
wdeg_data.plot.bar(
    x='White Degree', y='n_wdeg',
    legend=None,                 
    rot=45,                                     
    color='#d9e3f0'                             
);

# compares percentage of black people in work force with percent in a specific occupation
bdeg_data.plot.bar(
    x='Black Degree', y='n_bdeg',
    legend=None,                                
    rot=45,                                     
    color='#f47373'                             
);

# compares percentage of asian people in work force with percent in a specific occupation
adeg_data.plot.bar(
    x='Asian Degree', y='n_adeg',
    legend=None,                                
    rot=45,                                     
    color='#37d67a'                            
);

# compares percentage of hispanic people in work force with percent in a specific occupation
hdeg_data.plot.bar(
    x='Hispanic Degree', y='n_hdeg',
    legend=None,                                
    rot=45,                                     
    color='#ba68c8'                             
);
```

```python
male = pd.read_excel('../data/male_census_occupations.xlsx')
display(male)
```

```python
female = pd.read_excel('../data/female_census_occupations.xlsx')
display(female)
```

```python
#categorizing the percentage of a sex in the STEM field in comparison to that sex's percentage in earning STEM degrees

def categorize_male(percent):
    if percent < ((291791/589330)*100): return 'less than degree'
    elif percent == ((291791/589330)*100): return 'equal to degree'
    elif percent > ((291791/589330)*100): return 'more than degree'
    else: return np.nan

male['Male Degree'] = male['Total Percent'].apply(categorize_male)

def categorize_female(percent):
    if percent < ((297539/589330)*100): return 'less than degree'
    elif percent == ((297539/589330)*100): return 'equal to degree'
    elif percent > ((297539/589330)*100): return 'more than degree'
    else: return np.nan

female['Female Degree'] = female['Total Percent'].apply(categorize_female)
```

```python
#grouping the categorizations

mdeg_data = male.groupby('Male Degree').size().reset_index().rename(columns={0:'n_mdeg'})
display(mdeg_data)

fdeg_data = female.groupby('Female Degree').size().reset_index().rename(columns={0:'n_fdeg'})
display(fdeg_data)
```

```python
# compares percentage of white people in work force with percent in a specific occupation
mdeg_data.plot.bar(
    x='Male Degree', y='n_mdeg',
    legend=None,                 
    rot=45,                                     
    color='#d9e3f0'                             
);

# compares percentage of black people in work force with percent in a specific occupation
fdeg_data.plot.bar(
    x='Female Degree', y='n_fdeg',
    legend=None,                                
    rot=45,                                     
    color='#f47373'                             
);
```
